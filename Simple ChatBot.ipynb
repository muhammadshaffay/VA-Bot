```python
# Load Bot
import chat
bot = chat.Bot()
agent = bot.build_chain()
```

```python
# Chat
print("\t\t\tNOTE : Type 'Quit' To Leave Chat! \n\n")

while True:
    query = input("User : ")
    if query == 'Quit':
        print("Bot: See You Later!")
        break
    else:
        answer = agent.predict(user_input=query)
        print("Bot  :" + chat.display(answer) + "\n")
```
